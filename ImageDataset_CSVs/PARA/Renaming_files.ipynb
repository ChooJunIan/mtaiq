```python
image_path = 'E:/Dataset/PARA/imgs/'

for i in range(1, 447, 1):
    sessionId = 'session' + str(i)
    print(f'Renaming session ID: {i}...')
    session_path = image_path + sessionId + '/'
    j = 0
    for filename in os.listdir(session_path):
        j+=1
        source = session_path + filename
#         print(source)
        dest = sessionId + '_' + filename
#         print(dest)
        dest = image_path + dest
#         print(dest)
        os.rename(source, dest)
    
        
print("Session IDs merged with image names")
```
