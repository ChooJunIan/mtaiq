## We now begin storing the correlation scores of the all users' scores, to the mean score of the images, and see which users have strong correlation to the image mean score, by storing them in a DataFrame. 

```python
import math
# get all unique images
user_slice = para_images.loc[:, 'userId'].unique()

# dataframe for storing correlation scores
correlationScore = pd.DataFrame({'userId': pd.Series(dtype='str'),
                                'aestheticScore': pd.Series(dtype='float'),
                                'colorScore': pd.Series(dtype='float'),
                                'compositionScore': pd.Series(dtype='float'),
                                'contentPreference': pd.Series(dtype='float'),
                                'dofScore': pd.Series(dtype='float'),
                                'lightScore': pd.Series(dtype='float'),
                                'qualityScore': pd.Series(dtype='float'),
                                'willingnessToShare': pd.Series(dtype='float')})    

# loop to calculate correlation scores for each unique image
for user in range(len(user_slice)):
    sliced = all_user_score_to_image_average_score.loc[all_user_score_to_image_average_score['userId'] == user_slice[user]]
    temp = [user_slice[user], 
            round(sliced['aestheticScore_userScore'].corr(sliced['aestheticScore_imageMeanScore']), 2),
            round(sliced['colorScore_userScore'].corr(sliced['colorScore_imageMeanScore']), 2),
            round(sliced['compositionScore_userScore'].corr(sliced['compositionScore_imageMeanScore']), 2),
            round(sliced['contentPreference_userScore'].corr(sliced['contentPreference_imageMeanScore']), 2),
            round(sliced['dofScore_userScore'].corr(sliced['dofScore_imageMeanScore']), 2),
            round(sliced['lightScore_userScore'].corr(sliced['lightScore_imageMeanScore']), 2),
            round(sliced['qualityScore_userScore'].corr(sliced['qualityScore_imageMeanScore']), 2),
            round(sliced['willingnessToShare_userScore'].corr(sliced['willingnessToShare_imageMeanScore']), 2)]
    temp = pd.Series(temp).fillna(0).tolist()
    correlationScore.loc[len(correlationScore)]=temp
```

```python
import plotly.express as px

fig = px.box(correlationScore, y="aestheticScore", title="aestheticScore")
fig.show()
fig = px.box(correlationScore, y="colorScore", title="colorScore")
fig.show()
fig = px.box(correlationScore, y="compositionScore", title="compositionScore")
fig.show()
fig = px.box(correlationScore, y="contentPreference", title="contentPreference")
fig.show()
fig = px.box(correlationScore, y="dofScore", title="dofScore")
fig.show()
fig = px.box(correlationScore, y="lightScore", title="lightScore")
fig.show()
fig = px.box(correlationScore, y="qualityScore", title="qualityScore")
fig.show()
fig = px.box(correlationScore, y="willingnessToShare", title="willingnessToShare")
fig.show()
```

We take the lower fence for each scoring as the cut-off point for identifying the outliers only.

```python
outlier_df = correlationScore.loc[(correlationScore['aestheticScore'] < 0.59) | 
                                  (correlationScore['colorScore'] < 0.34) | 
                                  (correlationScore['compositionScore'] < 0.35) | 
                                  (correlationScore['contentPreference'] < 0.41) | 
                                  (correlationScore['dofScore'] < 0.33) | 
                                  (correlationScore['lightScore'] < 0.39) | 
                                  (correlationScore['qualityScore'] < 0.61) | 
                                  (correlationScore['willingnessToShare'] < 0.42)]
outlier_df.reset_index(drop=True, inplace=True)
```

```python
outlier_df
```

```python
outlier_count = []

for user in range(outlier_df.shape[0]):
    count = 0
    single_user = outlier_df.iloc[user]
    if (single_user['aestheticScore'] < 0.59):
        count += 1
    if (single_user['colorScore'] < 0.34):
        count += 1
    if (single_user['compositionScore'] < 0.35):
        count += 1
    if (single_user['contentPreference'] < 0.41):
        count += 1
    if (single_user['dofScore'] < 0.33):
        count += 1
    if (single_user['lightScore'] < 0.39):
        count += 1
    if (single_user['qualityScore'] < 0.61):
        count += 1    
    if (single_user['willingnessToShare'] < 0.42):
        count += 1
    outlier_count.append(count)
        
outlier_df['Outlier Count'] = outlier_count
```
